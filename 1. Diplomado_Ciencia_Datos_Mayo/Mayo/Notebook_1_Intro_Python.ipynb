Cuando trabajamos con números es importante redondear.

La función de redondeo en Python es:

$$\textbf{round(numero, # de decimales)}$$

- Si no se indica cuántos decimales entonces redondea al número entero más cercano

- Redondea a partir de 0.5 (estricto) en la cifra respectiva, es decir:
   
   - $round(3.125,2)=3.12$
   - $round(3.1251,2)=3.13$

- En el caso de redondear en $.5$, la función round redondea al número par más cercano (así es el funcionamiento en python).

````python
print(round(4.543))     # 5.0
print(round(4.43))      # 4
print(round(6.4689,3))  # 6.469
print(round(3.125,2))   # 3.12
````

````python
# observe lo siguiente: cuando es exactamente .5 : se envía al par más cercano.
print(round(4.545))
print(round(4.5))
print(round(5.5))
````

## <FONT SIZE=5 COLOR="green"> 8.2 Diccionarios en Python </FONT>

- Los diccionarios son una de las estructuras de datos más importantes y utilizadas en Python. Permiten almacenar información mediante pares de clave y valor, facilitando el acceso rápido y organizado a los datos.

$$clave : valor$$

- A diferencia de las listas, donde los elementos se acceden mediante posiciones numéricas, en los diccionarios cada valor está asociado a una clave única. Esto hace que la información sea más descriptiva, legible y fácil de manipular.

- Por ejemplo, en lugar de acceder al nombre de una persona usando una posición como persona[0], un diccionario permite utilizar una clave más clara como persona["nombre"].

- Los diccionarios son especialmente útiles cuando se trabaja con datos estructurados, configuraciones, registros, APIs, archivos JSON y aplicaciones de análisis de datos.

- Para acceder a los datos del diccionario se debe hacer por medio de la clave, que debe ser única.

- Una característica de los diccionarios es su representación en forma de llaves $\{$  $\}$

$$dicc = \{``clave_1" : datos_1 , ``clave_2" : datos_2 \dots ``clave_n" : datos_n \}$$

También existen otras formas de representación

$$dicc = dict([(``clave_1", datos_1) , (``clave_2", datos_2) \dots (``clave_1", datos_n)])$$

Veamos algunos ejemplos



**Crear un diccionario**

````python
# se puede crear usando {} así:
dicc = {"Nombre" : "Fabian",
         "Edad" : 24,
         "Sexo" : "Masculino",
         "Cédula" : 80987656}
print(dicc)
````

````python
# se puede crear usando dict() así:
dicc2 = dict(Nombre = "Fabian",
          Edad = 24,
          Sexo = "Masculino",
          Cédula = 80987656)
print(dicc2)
````

````python
# se puede crear usando la instrucción dict([])
dicc1 = dict([("Nombre", "fabian"),
           ("Edad", 24),
           ("Sexo", "Masculino"),
           ("Cédula", 80097865)])
print(dicc1)
````

Las claves no necesariamente deben tener un solo elemento

````python
dicc3 = dict(Nombre = ["fabian", "andres", "carlos"],
          Edad = [23,34,45])
print(dicc3)
````

````python
# se puede crear usando {} así:
dicc = {"Nombre" : ["Fabian", "Carlos", "Ana", "Oscar"],
         "Edad" : [24,30],
         "Sexo" : ["Masculino", "Masculino","Femenino"],
         "Cédula" : [80987656, 333333, 11111]}
print(dicc)
````

**Acceder a los datos del diccionario**

- Para acceder a un elemento del diccionario usamos la instrucción:

$$dicc["clave"]$$

Veamos algunos ejemplos.

````python
# utilizamos las claves para acceder al valor.
dicc["Sexo"]
````

````python
# {'Nombre': 'Fabian', 'Edad': 24, 'Sexo': 'Masculino', 'Cédula': 80987656}
# accedemos a la clave "nombre"
print(dicc["Nombre"])
````

````python
# {'Nombre': ['fabian', 'andres', 'carlos'], 'Edad': [23, 34, 45]}
# accedemos a la clave "nombre"
print(dicc3["Nombre"])
````
