## <FONT SIZE=4 COLOR="magenta"> 1.1 Ordenar según una variable </FONT>

Para ordenar por una variable usamos la instrucción

$$datos.sort\_values(variable, ascending = False) $$

se usa:

- *ascending=True* :  ordena de menor a mayor. (default)
- *ascending=False* : ordena de mayor a menor.

```python
# ordenamos por la variable precio de mayor a menor
aptos.sort_values("precio" , ascending =  False)
```

```python
# ordenamos por la variable precio de menor a mayor
aptos.sort_values("precio" , ascending =  True)
```

```python
# default ordena de menor a mayor por la variable ubicación
# lo hace por orden alfabético
aptos.sort_values("ubicacion", ascending = True)
```

```python
aptos.columns
```

## <FONT SIZE=4 COLOR="magenta"> 1.2 Renombrar una variable </FONT>

Podemos renombrar una columna de la siguiente forma usando la instrucción ***rename***

$$ datos.rename(columns=\{key : value, key : value\}, inplace=True) $$

 - Para cambiar el nombre de una variable debemos usar un diccionario donde las claves son las variables actuales y los valores son los nombres nuevos de las variables.
 - Recordemos que la instrucción **inplace** se usa para determinar si cambiamos el DataFrame original.

```python
# es recomendable hacer una copia
aptos2 = aptos.copy()
```

```python
# cambiaremos el nombre de administracion por admon
aptos2.rename(columns={"administracion" : "admon"}, inplace = True)
```

```python
# ahora sí quedó sobre datos2.
aptos2.head()
```

```python
# ejemplo
homicidios.columns
```

```python
# cambiar Estado civil por e.civil y Móvil Victima por m.victima
homicidios2 = homicidios.copy()
homicidios2.rename(columns={"Estado civil" : "e.civil" , "Móvil Victima" : "m.victima"}, inplace = True)
```

```python
homicidios2.columns
```

## <FONT SIZE=4 COLOR="magenta"> 1.3 Eliminar filas y columnas</FONT>

Para eliminar una columna o una fila usamos la instrucción .drop

$$datos.drop(``nombre", axis = 1, inplace = T \, or \, F)$$

- *axis = 1* : es por columna
- *axis = 0* : es por  fila
- *inplace = T* : sobreescribe en el *df* original.
- *inplace = F* : no sobreescribe. Mantiene el original

```python
aptos2.head()
```

```python
# Eliminar la variable "Unnamed: 0"
aptos2.drop("Unnamed: 0", axis = 1 , inplace = True)
```

```python
aptos2.head()
```

```python
homicidios2.columns
```

```python
homicidios2.head()
```

```python
aptos.columns
```

```python
aptos2.drop([4,6], axis =0)
```

```python
aptos2.head()
```

## <FONT SIZE=4 COLOR="magenta"> 1.4 Agregar una variable </FONT>

Para agregar una variable usamos la instrucción

$$datos[``\text{nueva variable}"] = lista $$

```python
aptos2.head()
```

## <FONT SIZE=4 COLOR="magenta"> 1.5 Agrupaciones: ***groupby()*** </FONT>

Podemos hacer operaciones con las agrupaciones, así:

$$datos.groupby(``Variable")[``\text{segunda variable}"].funcion()$$

Por ejemplo, supongamos que queremos agrupar por escolaridad y calcular el promedio de las edades de cada uno de los grupos.

<center><img src="https://github.com/Fabian830348/cursos/blob/master/intro/groupby.png?raw=true" alt="centered image" width="450" height="350"></center> <center><figcaption> <FONT SIZE=1 COLOR="black"> Fuente: Elaboración propia  </FONT> <figcaption></center>


```python
aptos2.groupby("ubicacion")
```

```python
# calcula la media del precio en cada ubicación
aptos2.groupby("ubicacion")["precio"].mean()
```

```python
# calcular las estadísticas del precio en cada ubicación
aptos2.groupby("ubicacion")["precio"].describe()
```

```python
aptos2.parqueadero.unique()
```

```python
# si agrupamos por ubicacion, y parqueadero quiero saber el valor máximo del precio
aptos2.groupby(["ubicacion" , "parqueadero"]).precio.max()
```

```python
aptos.groupby(["ubicacion", "parqueadero"]).precio.min()
```

# <FONT SIZE=5 COLOR="green"> 2. Filtros en un DataFrame </FONT>

En esta sección revisaremos algunos filtros en un dataframe. Para esto, generaremos algunas preguntas y buscaremos responderlas con los filtros.

De esta manera podemos seguir haciendo filtros y respondiendo algunas preguntas. Se debe tener en cuenta:

- Operadores de comparación :

    - **>** : mayor que
    - **<** : menor que
    - **>=**  : mayor o igual
    - **<=** : menor o igual
    - **!=** (diferente)

- Operadores lógicos:

    - **&**  : y
    - **|**  : o

<FONT SIZE=3 COLOR="blue"> Pregunta 1: ¿Cuántos apartamentos tienen precio mayor a 100 millones? </FONT>

```python
aptos2.head()
```

```python
sum(aptos2.precio > 400)
```

```python
# apartamentos de mas de 400 millones
aptos2[aptos2.precio > 400]
```

```python
# cuántos apartamentos de mas de 400 millones
len(aptos2[aptos2.precio > 400])
```

<FONT SIZE=3 COLOR="blue"> Pregunta 2: ¿Cuántos apartamentos del poblado tienen precio mayor a 200 millones? </FONT>

```python
aptos2[(aptos2["precio"]==420) & (aptos2["ubicacion"] == "poblado")]
```

```python
len(aptos2[(aptos2["precio"]==420) & (aptos2["ubicacion"] == "poblado")])
```

<FONT SIZE=3 COLOR="blue"> Pregunta 3: ¿Cuántos apartamentos del sector laureles, tienen más de 50 metros cuadrados y no cuentan con parqueadero? </FONT>

```python
len(aptos2[(aptos2["mt2"]>50) & (aptos2["ubicacion"]=="laureles") & (aptos2["parqueadero"]=="no")])
```
