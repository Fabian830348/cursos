# <FONT SIZE=5 COLOR="purple"> 2. Funciones en Python </FONT>

- Las funciones son bloques de código reutilizables que permiten organizar y estructurar programas de manera eficiente.

- Su principal objetivo es encapsular tareas específicas, facilitando la lectura, el mantenimiento y la reutilización del código.

- En Python, las funciones se definen mediante la palabra clave *def*
y pueden recibir parámetros de entrada, realizar operaciones y devolver resultados.

```python
def function(x,y,x..):
  .....
  return ...  
```
- El uso de funciones contribuye a reducir la duplicación de código, mejorar la modularidad de los programas y simplificar la resolución de problemas complejos mediante su división en tareas más pequeñas.-

- Favorecen la claridad del código y permiten desarrollar aplicaciones más robustas y fáciles de depurar.

En esta sección se presentan los conceptos básicos relacionados con la definición, uso y aplicación de funciones en Python, así como algunos ejemplos prácticos.

````python
# ejemplo básico de función sin parámetros
def saludo1():
  print("Hola cómo están. Esto es un saludo")
````

cada vez que escribamos *saludo()*

````python
saludo1()
````

````python
# ejemplo básico de función con parámetros
def saludo2(nombre):
  print("Hola" , nombre, "Cómo estás?")
````

````python
saludo2("Fabian")
````

**Formas diferentes del print()**

Para imprimir variables dentro del *print* hay varias alternativas

- **f-strings**

- **.format()**

**método f"{variable}"**

````python
# método f"{variable}"
def saludo3(nombre):
  print(f"Hola, {nombre} espero que estes bien")
````

````python
saludo3("Carolina")
````

**método "{}".format(variable)**

````python
def saludo4(nombre):
    print("Hola {}, me alegra mucho verte".format(nombre))
````

````python
saludo4("Carlos")
````

Las funciones pueden tener varios parámetros

````python
def suma1(x,y):
  print(x+y)

suma1(10,89)
````

````python
# veamos el valor de x
print(x)
````

No guarda la información!!

````python
suma2(10,89)
````

````python
#hacemos lo mismo
x=suma2(10,89)
print(x)
````

Podemos pensar como una calculadora.

- **print**:

  - la calculadora muestra el resultado en la pantalla.
  - cuando se quiere mostrar únicamente la información

- **return**:

  - la calculadora te entrega el resultado para que puedas usarlo en otro cálculo.
  - cuando queremos un resultados para seguir trabajando con él.

````python
def exploracion(datos):
    print("Primeros 5 registros \n")
    display(datos.head())

    print("\n Últimos 5 registros \n")
    display(datos.tail())

    print("\n Tamaño de los datos \n")
    print(datos.shape)
    print(f"Filas: {datos.shape[0]}")
    print(f"Columnas: {datos.shape[1]}")

    print("\n Nombres de las columnas \n")
    print(datos.columns)

    print("\n Información de los datos \n")
    print(datos.info())

    print("\n Estadísticas descriptivas \n")
    display(datos.describe())

    print("\n Valores faltantes \n")
    display(datos.isna().sum())
````

````python
exploracion(credito)
````

````python
def dimensiones(df):
    return pd.DataFrame({
        "Filas": [df.shape[0]],
        "Columnas": [df.shape[1]]
    })
````

````python
dimensiones(credito)
````

````python
def tipos(df):
    return pd.DataFrame({
        "Columna": df.columns,
        "Tipo": df.dtypes.values
    })
````

````python
tipos(credito)
````

````python
def valores_unicos(df, columna):
    return pd.DataFrame({
        "Valores únicos": df[columna].unique()
    })
````
