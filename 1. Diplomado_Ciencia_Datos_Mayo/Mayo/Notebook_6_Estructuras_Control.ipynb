# <FONT SIZE=5 COLOR="purple"> 1. Estructuras de Control </FONT>

Una **estructura de control** es un concepto básico de la programación (especialmente en el campo de la ciencias de la computación) que sirve para decidir cómo se ejecutan las instrucciones de un programa.

En pocas palabras:

- Indica en qué orden se ejecuta el código, si se repite o si se toman decisiones.

- Estas estructuras de permite modificar el orden en el cual se ejecutan las instrucciones de un algoritmo.

##  <FONT SIZE=5 COLOR="green"> Tipos de estructuras de control </FONT>

**1. Secuencial**

Es la forma más simple: las instrucciones se ejecutan una tras otra, en orden.

```python
leer número
sumar 5
mostrar resultado
```

**2. Condicional (decisión)**

Permite que el programa tome decisiones dependiendo de una condición.

```python
si (edad >= 18)
    mostrar "Eres mayor de edad"
si no
    mostrar "Eres menor de edad"
```

**3. Repetitiva (bucles)**

Sirve para repetir instrucciones varias veces.

```python
mientras (contador < 5)
    mostrar contador
    contador = contador + 1
```

**¿Para qué sirven?**

Las estructuras de control permiten que un programa:

- Tome decisiones
- Repita acciones
- Controle el flujo del programa

Sin ellas, los programas solo ejecutarían instrucciones de forma lineal, sin lógica.

Ejemplo

```python
edad_maria = 56
edad_alberto = 47

# Si María es mayor en edad que Alberto entonces:
if (edad_maria > edad_alberto):
  print('María tiene más años que Alberto.')
# De lo contrario:
else:
  print('Alberto tiene mas años que María.')

```

# <FONT SIZE=5 COLOR="purple"> 2. Condicional if</FONT>

El condicional **if** es una estructura de control que permite que un programa tome decisiones. Evalúa una condición y, si esta es verdadera, ejecuta un bloque de instrucciones. Si la condición es falsa, ese bloque no se ejecuta (o se ejecuta otro bloque si se utiliza **else**).

```python
if condición:
    instrucciones
```

Ejemplo

```python
edad = 20
if edad >= 18:
    print("Eres mayor de edad")
```

**Uso de else**

Cuando queremos ejecutar una acción diferente si la condición es falsa, utilizamos *else:*

```python
edad = 16

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

Las estructuras condicionales permiten que los programas reaccionen a diferentes situaciones y tomen decisiones de manera automática. Gracias al **if**, es posible validar datos, controlar accesos, comparar valores y adaptar el comportamiento del programa según las condiciones que se presenten.








````
# ejemplo 1
a = 15
b = 10

if a > b:
    print("a es mayor que b")
````

````
# ejemplo 2
edad = 16

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
````

````
#ejemplo 3
temperatura = 25

if temperatura > 20:
    print("Clima agradable")
````

# <FONT SIZE=5 COLOR="purple"> 2. Cíclica *for* </FONT>

El ciclo *for* es una estructura de control repetitiva que permite ejecutar un bloque de código varias veces. Se utiliza cuando conocemos cuántas veces queremos repetir una acción o cuando queremos recorrer los elementos de una colección (lista, cadena, DataFrame, etc.).

```python
for variable in secuencia:
    instrucciones
```






````
# ejemplo 1
frutas = ['manzana', 'pera', 'uva']

for fruta in frutas:
    print(fruta)
````

````
# ejemplo 2
for i in range(5):
    print(i)
````

````
# ejemplo 3
suma = 0

for i in range(1, 6):
    suma = suma + i

print(suma)
````

````
# ejemplo 4
for letra in "Python":
    print(letra)
````

# <FONT SIZE=5 COLOR="purple"> 2. El ciclo *while* </FONT>

El ciclo *while* es una estructura de control repetitiva que permite ejecutar un bloque de instrucciones mientras una condición determinada sea verdadera.

A diferencia del ciclo for, que suele utilizarse cuando se conoce de antemano el número de iteraciones, el while resulta especialmente útil cuando no se sabe cuántas veces debe repetirse una acción y esta depende del cumplimiento de una condición.

En cada iteración, el programa evalúa la condición y, si continúa siendo verdadera, ejecuta nuevamente el bloque de código.

Por esta razón, es importante que la condición cambie en algún momento dentro del ciclo para evitar repeticiones infinitas. El uso adecuado de while permite automatizar tareas, controlar procesos iterativos y gestionar situaciones donde la repetición depende de eventos o resultados que se producen durante la ejecución del programa.

```python
while condición:
    instrucciones
```



````
# Ejemplo 1
contador = 1

while contador <= 3:
    print(contador)
    contador += 1
else:
    print("El ciclo ha terminado")
````

````
# ejemplo 2
while i < len(aptos):

    if aptos.loc[i, "precio"] > 600:
        print("Valor atípico encontrado:")
        print(aptos.loc[i])
        break

    i += 1
````
