# <FONT SIZE=4 COLOR="green"> 2. Histogramas </FONT>

- El histograma se usa para variables de tipo cuantitativo continuo o discreto con muchas mediciones diferentes (como la edad en ciertos casos).

- En el eje $x$ se identifican las clases (intervalos) de la variable.

- En el eje $y$ se identifica la frecuencia de cada una de las clases.

- Las barras en el histograma quedan contiguas y además se utiliza un solo color en estas.

```python
# tabla de frecuencias
aptos.precio.value_counts(bins =10)
```

## <FONT SIZE=4 COLOR="green"> 2.1 Histograma con Matplotlib </FONT>

```python
# histograma de la variable precio con matplotlib
plt.hist(aptos["precio"])
plt.show()
```

```python
# Histograma de la variable precio
plt.hist(
    aptos["precio"],
    color="skyblue",      # color de relleno
    edgecolor="black",    # color del borde
    linewidth=1.2         # grosor del borde
)

plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.title("Histograma de la variable precio")

plt.show()
```

## <FONT SIZE=4 COLOR="green"> 2.1 Histograma con Plotly Express </FONT>

```python
# solo colocar la variable
px.histogram(aptos.precio)
```

```python
# histograma de la variable precio con plotly especificando
px.histogram(aptos,
             x = "precio")
```

```python
# gráfico de histograma con edición
fig = px.histogram(aptos,
                   x="precio",
                   nbins = 20)

fig.update_traces(
    marker_color="steelblue",   # color de las barras
    marker_line_color="black",  # color del borde
    marker_line_width=1.5       # grosor del borde
)

fig.show()
```

# <FONT SIZE=4 COLOR="green"> 3. Boxplot (gráfico de cajas y bigotes)</FONT>

- El gráfico de caja, boxplot del inglés, fue ideado por Jhon Tukey en 1977 y es una representación estadística que resalta principalmente aspectos de la distribución de los datos u observaciones por medio de las medidas de posición.

- Los diagramas de caja brindan información valiosa que no se limpia en un histograma o en un polígono de frecuencias.

A continuación vamos a describir cada uno de los elementos que tiene un boxplot

**La caja rectangular** : el gráfico contiene una caja rectangular. Es importante indicar que esta caja no tiene coordenadas cartesianas en el eje $X$  y las alturas en el eje $y$ no representa una escala de frecuencias. Las lineas horizontales en el diagrama de cajas describen el primer y tercer cuartil. $Q_1$ y $Q_3$ respectivamente.

**Mediana** : es una línea que se muestra dentro de la caja rectangular y hace referencia al valor que divide los datos en dos partes iguales. Recuerde que la mediana es igual a $Q_2$.

**Bigotes** : son líneas que salen a los costados de la caja y se usan como referencia para ubicar las observaciones que estan por fuera del $50 \%$ central de distribucion. Los bigotes nos dan información sobre la existencia de valores atípicos (*outliers*).

## <FONT SIZE=4 COLOR="green"> 3.1 Boxplot con Matplolib </FONT>

```python
# boxplot de la variable precio con matplotlib
plt.boxplot(aptos.precio)
plt.show()
```

```python
fig = plt.boxplot(
    aptos["precio"],
    patch_artist=True,
    boxprops=dict(facecolor="skyblue")
)
plt.show()
```

## <FONT SIZE=4 COLOR="green"> 3.2 Boxplot con Plotly Express </FONT>

```python
# boxplot de la variable precio con plotly
px.box(aptos.precio)
```

```python
# boxplot de la variable precio con plotly
fig = px.box(aptos,
             y="precio",
             color_discrete_sequence=["green"]
)

fig.show()
```

# <FONT SIZE=4 COLOR="green"> 4. Gráfico de Violin </FONT>

El gráfico de violín es una herramienta de visualización utilizada para analizar la distribución de una variable numérica. Combina las características de un diagrama de caja (boxplot) con una estimación de densidad, permitiendo identificar la tendencia central, la dispersión de los datos y las regiones donde se concentran con mayor frecuencia las observaciones. Su forma facilita la detección de asimetrías, multimodalidad y posibles patrones que podrían no apreciarse claramente mediante otras representaciones gráficas, proporcionando una visión más detallada de la distribución de los datos.


```python
# gráfico de violin
px.violin(aptos,
          y="precio",
          color_discrete_sequence=["purple"])
```

# <FONT SIZE=4 COLOR="green"> 6. Otros gráficos </FONT>

**Boxplot por cada categoria de una variable cualitativa**

```python
boxplot = px.box(aptos,                                 # nombre de los datos
       y = "precio",                                    # variable cuantitativa
       x = "ubicacion",                                 # variable cualitativa o categórica
       title = "Boxplot variable precio vs ubicación",  # título del gráfico
       color =  "ubicacion")                            # color por la variable categórica

boxplot.write_html("box.html") # para guardar la imagen
```

**Diagrama de dispersión con cruces de otras variables**

```python
px.scatter(aptos,
           x= "mt2",
           y= "precio",
           color = "ubicacion")          # categórica
```

```python
px.scatter(aptos,
           x= "mt2",
           y= "precio",
           color = "ubicacion",
           size = "administracion")
```

```python
px.scatter(aptos,
           x= "mt2",
           y= "precio",
           size = "administracion",   # el tamaño: variable numérica
           color = "ubicacion",       # categórica
           symbol= "balcon")          # categórica
```

**Gráfico con más edición**

```python
# diagrama de dispersión con color de acuerdo con categorías de una variable categórica
fig = px.scatter(aptos,                                     # datos
                 x= "precio",                               # variable 1
                 y= "mt2",                                  # variable 2
                 color = "ubicacion")                       # color por categoría

fig.update_layout(yaxis_title='precio',                              # título del eje y
                  xaxis_title = "metros cuadrados",                  # título del eje x
                  title = "Diagrama de dispersión: precio vs mt2"    # título del gráfico
)

fig.show()
```

**Gráfico polar**

```python
# gráfico polar frecuencias
fig = px.line_polar(aptos,
                    r = 'mt2',
                    theta = 'ubicacion',
                    line_close = True,
                    color_discrete_sequence = ['magenta'])

fig.update_traces(fill = "toself")

fig.show()
```

```python
# mapa de calor y guardar en html
fig = px.density_heatmap(aptos,
                         x="precio",
                         y="ubicacion")
fig.show()
fig.write_html("mapa_de_calor.html")
```
