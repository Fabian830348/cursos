# <FONT SIZE=5 COLOR="blue"> 1. Contexto del problema </FONT>

Este conjunto de datos representa la operación diaria de un sistema energético híbrido conectado a red, que integra generación renovable (solar y eólica), almacenamiento en baterías y demanda eléctrica variable. Las variables describen tanto condiciones ambientales (irradiancia solar, temperatura, humedad, viento) como variables operativas del sistema eléctrico (demanda, voltaje de red, eficiencia del inversor, estado de carga de la batería y tarifa). El objetivo del conjunto de datos es identificar la ocurrencia de un evento de estrés en la red eléctrica, el cual puede asociarse a escenarios de alta demanda, bajo soporte del almacenamiento, caídas de voltaje o pérdida de eficiencia, situaciones críticas para la operación segura y eficiente del sistema. Este contexto permite aplicar técnicas de exploración de datos y clasificación con machine learning para apoyar la toma de decisiones operativas, la gestión energética y la prevención de fallas en infraestructuras energéticas modernas.

 - 2000 registros

 - 10 variables independientes

 - 1 variable objetivo dicotómica (grid_stress_event)

| **Variable**           | **Tipo**             | **Unidad / Valores** | **Descripción**                                                                                                              |
| ---------------------- | -------------------- | -------------------- | ---------------------------------------------------------------------------------------------------------------------------- |
| `solar_irradiance_wm2` | Numérica continua    | W/m²                 | Irradiancia solar incidente en el sitio de generación fotovoltaica. Indica el potencial instantáneo de generación solar.     |
| `wind_speed_ms`        | Numérica continua    | m/s                  | Velocidad del viento medida en el punto de generación eólica, asociada al potencial de producción de energía eólica.         |
| `ambient_temp_c`       | Numérica continua    | °C                   | Temperatura ambiente del entorno operativo del sistema energético.                                                           |
| `humidity_pct`         | Numérica continua    | %                    | Humedad relativa del ambiente, que puede afectar el desempeño térmico de equipos eléctricos.                                 |
| `panel_temp_c`         | Numérica continua    | °C                   | Temperatura del panel fotovoltaico, influenciada por la irradiancia solar y la temperatura ambiente.                         |
| `inverter_efficiency`  | Numérica continua    | 0 – 1                | Eficiencia del inversor de potencia, que convierte corriente continua en alterna. Valores bajos indican pérdidas operativas. |

| `battery_soc_pct`      | Numérica continua    | %                    | Estado de carga (SoC) de la batería del sistema de almacenamiento energético.                                                |
| `tariff_rate_usd_kwh`  | Numérica continua    | USD/kWh              | Tarifa de energía eléctrica aplicada en el periodo analizado.                                                                |
| `grid_stress_event`    | Categórica (binaria) | 0 / 1                | **Variable objetivo**: indica si ocurre un evento de estrés en la red eléctrica (1) o una operación normal (0).              |
