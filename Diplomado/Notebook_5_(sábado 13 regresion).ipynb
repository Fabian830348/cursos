# <FONT SIZE=5 COLOR="Purple"> 4. Modelo KNN </FONT>

```python
# 1. identificar las variables
y = regresion["AhorroPotencial_USD"]
X = regresion.drop("AhorroPotencial_USD", axis=1)

# 2. Conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state = 123,
                                                    test_size = 0.3)

# 3. Escalar los datos
escalar = StandardScaler()
X_train_s = escalar.fit_transform(X_train)
X_test_s = escalar.transform(X_test)

# 4. Definir el modelo de Regresión logística
modelo_knn = KNeighborsRegressor(n_neighbors=5)

# 5. Entrenar el modelo
modelo_knn.fit(X_train_s,y_train)

# 6. Predicciones
y_pred = modelo_knn.predict(X_test_s)

# 7. Calcular la métricas
print("R2",r2_score(y_test, y_pred))                             # coeficiente de determinación
print('MSE', mean_squared_error(y_test, y_pred))                 # error cuadrático medio
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred)))        # raíz error cuadrático medio
print("MAE:", mean_absolute_error(y_test, y_pred))
```

1. $R^2$: Coeficiente de determinación

  - Indica qué tan bien el modelo explica los datos reales.

  - Se interpreta:
     
     - Va de 0 a 1 (a veces negativo)

     - Mientras más cerca de 1, mejor el modelo


2. $MAE$  Error Absoluto Medio

  - El promedio de los errores entre valores reales y predichos, sin signos negativos.

  - Se interpreta:
     
     - Está en las mismas unidades de la variable objetivo

     - Mientras pequeño, mejor el modelo

3. $MSE$  Error Cuadrático Medio

  - El promedio de los errores elevados al cuadrado.

  - Se interpreta:
     
     - Penaliza más los errores grandes

     - No está en unidades interpretables directamente

     - Cuánto castiga el modelo los errores grandes.

3. $RMSE$  Raíz del Error Cuadrático Medio

  - La raíz cuadrada del MSE.

  - Se interpreta:
     
     - Vuelve a las mismas unidades de la variable objetivo

     - Sensible a errores grandes

     - Qué tan grande es el error típico del modelo.


# <FONT SIZE=5 COLOR="Purple"> 5. Árboles de Decisión </FONT>

```python
# 1. identificar las variables
y = regresion["AhorroPotencial_USD"]
X = regresion.drop("AhorroPotencial_USD", axis=1)

# 2. Conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state = 123,
                                                    test_size = 0.3)

# 4. Definir el modelo de Regresión logística
modelo_tree = DecisionTreeRegressor()

# 5. Entrenar el modelo
modelo_tree.fit(X_train,y_train)

# 6. Predicciones y matriz de confusión
y_pred = modelo_tree.predict(X_test)

# 7. Calcular la métricas
print("R2",r2_score(y_test, y_pred))                             # coeficiente de determinación
print('MSE', mean_squared_error(y_test, y_pred))                 # error cuadrático medio
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred)))        # raíz error cuadrático medio
print("MAE:", mean_absolute_error(y_test, y_pred))
```

# <FONT SIZE=5 COLOR="Purple"> 6. Random Forest </FONT>

```python
# 1. identificar las variables
y = regresion["AhorroPotencial_USD"]
X = regresion.drop("AhorroPotencial_USD", axis=1)

# 2. Conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state = 123,
                                                    test_size = 0.3)

# 3. Definir el modelo de Regresión logística
modelo_rf = RandomForestRegressor()

# 4. Entrenar el modelo
modelo_rf.fit(X_train,y_train)

# 5. Predicciones
y_pred = modelo_rf.predict(X_test)

# 6. Calcular la métricas
print("R2",r2_score(y_test, y_pred))                             # coeficiente de determinación
print('MSE', mean_squared_error(y_test, y_pred))                 # error cuadrático medio
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred)))        # raíz error cuadrático medio
print("MAE:", mean_absolute_error(y_test, y_pred))
```

# <FONT SIZE=5 COLOR="Purple"> 7. Validar el modelo </FONT>

```python
from sklearn.model_selection import cross_val_score

# Validación cruzada k-fold
resultados = cross_val_score(modelo_rf,                          # modelo
                             X,                                  # predictores
                             y,                                  # variable objetivo
                             cv=10,                              # 10 folds
                             scoring="neg_mean_absolute_error")

# Convertir a positivo
rmse_cv = -resultados.mean()

print("MAE promedio (CV 10-fold):", rmse_cv)
```
