# <FONT SIZE=5 COLOR="purple"> 2. Tokenización </FONT>

- Es el proceso de dividir las cadenas de texto de un documento en piezas o unidades más pequeñas que se denominan *tokens*.

- Los *tokens* pueden ser palabras, subpalabras o incluso caracteres.

- La librería ***NLTK*** tiene funciones para hacer la tokenización de textos.

```python
# librerías para procesar
import numpy as np
import pandas as pd

# expresiones regulares
import re

# librerías para graficar
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# librería de PLN en español
import nltk
from nltk.corpus import stopwords
#stopwords_sp = stopwords.words('spanish')

# Count vectorizer usando NLTK
from sklearn.feature_extraction.text     import CountVectorizer
from sklearn.feature_extraction.text     import TfidfVectorizer

# Para el algoritmo k-means
from sklearn.cluster             import KMeans

# Algunos elementos de preprocesamiento: escalamiento y selección de k en k-means
from sklearn.preprocessing      import StandardScaler, MinMaxScaler
from sklearn.metrics            import silhouette_samples, silhouette_score

# Para los algotitmos de machine learning
from sklearn.decomposition      import PCA
from sklearn.neighbors          import KNeighborsClassifier
from sklearn.neighbors          import NearestNeighbors

from sklearn.metrics            import pairwise_distances_argmin_min

#Para omitir los warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# recordemos la función de preprocesado
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub('á', 'a', texto)
    texto = re.sub('é', 'e', texto)
    texto = re.sub('í', 'i', texto)
    texto = re.sub('ó', 'o', texto)
    texto = re.sub('ú', 'u', texto)
    texto = re.sub("[^a-züéáíóúñ ]", " ", texto)
    texto = [palabra for palabra in texto.split() if (len(palabra)>3 and palabra not in stopwords_sp) | (palabra == "no")]
    texto = " ".join(texto)
    return texto
```

```python
stopwords_sp = stopwords.words('spanish')
```

```python
comentario_limpio = [pre_procesado(comentario)  for comentario in datos.comentario]
```

```python
datos["comentario_limpio"] = comentario_limpio
```

```python
datos.head()
```

```python
# aplicar el tfidf vectorizar. Recordar: sparse matrix
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(datos.comentario_limpio)
tfidf_matrix = pd.DataFrame(tfidf.toarray(), columns = tfidf_vect.get_feature_names_out())
tfidf_matrix
```

```python
# veamos el método del codo
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()                                          # definimos el modelo
visualizer = KElbowVisualizer(model, k=(1,10))            # definimos el visualizador
visualizer.fit(tfidf_matrix)                              # Entrenamos con los datos
visualizer.show()
```

```python
k = 4
model = KMeans(n_clusters=k,
               random_state=42)
model.fit(tfidf_matrix)
```

```python
# pegamos los clusters
datos["cluster"] = model.labels_
datos[["comentario" , "cluster"]]
```

```python
datos['cluster'].value_counts(normalize=True)
```

```python
vecinos = NearestNeighbors(n_neighbors=5, metric='euclidean').fit(tfidf_matrix.values)

freq_cluster = datos["cluster"].value_counts()
freqrel_cluster = datos['cluster'].value_counts(normalize=True)

centroides = model.cluster_centers_

palabras = tfidf_vect.get_feature_names_out()
order_centroides = centroides.argsort()[:, ::-1]

for i in range(k):
    print(f"Cluster # {i}")
    print(f"Tiene {freq_cluster[i]} registros ({freqrel_cluster[i]:.2%} de los datos)")
    print()

    print("Palabras más representativas")
    for ind in order_centroides[i][:20]:
        print(palabras[ind], end=" ")
    print()
    print()

    print(f"Documentos más representativas")
    for vecino in datos.iloc[vecinos.kneighbors([centroides[i]])[1][0]].comentario.values:
        print(f" *** {vecino}")
        print()

    print()
    print("********************************************")
    print()
```

```python
import re
import pandas as pd
from pathlib import Path

# --------- 1) Lector de PDF (pdfplumber -> PyPDF2 fallback) ----------
def read_pdf_text(pdf_path: str) -> str:
    text = ""
    try:
        import pdfplumber
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                text += page.extract_text() or ""
    except Exception:
        try:
            import PyPDF2
            with open(pdf_path, "rb") as f:
                reader = PyPDF2.PdfReader(f)
                for page in reader.pages:
                    text += page.extract_text() or ""
        except Exception as e:
            raise RuntimeError(f"No pude leer el PDF. Instala pdfplumber o PyPDF2. Error: {e}")
    return text

# --------- 2) Parsers (expresiones regulares sobre el texto) ----------
def parse_historia(text: str) -> dict:
    # Normaliza espacios
    t = re.sub(r"[ \t]+", " ", text)
    t = re.sub(r"\u00A0", " ", t)  # non-breaking space

    def find(pattern, flags=re.IGNORECASE):
        m = re.search(pattern, t, flags)
        return m.group(1).strip() if m else ""

    data = {}
    data["institucion"] = find(r"^(INSTITUCIÓN DE SALUD [A-ZÁÉÍÓÚÑ ]+)", flags=re.IGNORECASE|re.MULTILINE)
    data["tipo_doc"] = "Historia clínica simulada"

    data["nombre"] = find(r"Nombre:\s*([^\|]+)")
    data["documento"] = find(r"Documento:\s*([^\n|]+)")
    data["fecha_nacimiento"] = find(r"Fecha de nacimiento:\s*([^\(]+)").strip()
    data["edad"] = find(r"\(\s*(\d+)\s*años?\s*\)")
    data["sexo"] = find(r"Sexo:\s*([MF])")
    data["eps"] = find(r"EPS:\s*([^\|]+)")
    data["regimen"] = find(r"Régimen:\s*([^\n|]+)")
    data["fecha_atencion"] = find(r"Fecha y hora de atención:\s*([^\|]+)")
    data["numero_historia"] = find(r"(?:Número|Número de historia):\s*([A-Z0-9\-]+)")

    # Signos vitales (en una sola línea)
    sv = find(r"Signos vitales:\s*([^ \n].+)")
    data["signos_vitales"] = sv
    data["ta"] = find(r"TA\s*([0-9]{2,3}\/[0-9]{2,3})")
    data["fc"] = find(r"FC\s*([0-9]{2,3})")
    data["fr"] = find(r"FR\s*([0-9]{1,2})")
    data["temp"] = find(r"Temp\s*([0-9]{1,2}\.[0-9])")
    data["spo2"] = find(r"SpO2\s*([0-9]{2,3})")

    # Motivo consulta
    data["motivo_consulta"] = find(r"2\. MOTIVO DE CONSULTA\s*“?([^”\n]+)")

    # Diagnósticos: intenta capturar códigos CIE-10 como X00, R50.9, etc.
    dx_block = find(r"7\. IMPRESIÓN DIAGNÓSTICA\s*([^\n]+)")
    if not dx_block:
        dx_block = find(r"7\.\s*IMPRESIÓN DIAGNÓSTICA\s*([^\n]+)")
    # Extrae tuplas (codigo – descripcion)
    dx_list = re.findall(r"([A-Z][0-9][0-9](?:\.[0-9])?)\s*[\-–]\s*([^;]+)", t, flags=re.IGNORECASE)
    diagnosticos = [{"codigo": c.strip().upper(), "descripcion": d.strip()} for c, d in dx_list]

    # Plan (líneas principales)
    plan_match = re.search(r"8\.\s*PLAN\s*[\-–]\s*CONDUCTA\s*(.+?)\n9\.", t, flags=re.IGNORECASE|re.DOTALL)
    plan_items = []
    if plan_match:
        plan_text = plan_match.group(1)
        for line in plan_text.split("\n"):
            line = line.strip(" •-—\t ")
            if len(line) > 3:
                plan_items.append(line)

    # Profesional
    prof_line = find(r"10\.\s*PROFESIONAL\s*([^\n]+)")
    if not prof_line:
        prof_line = find(r"11\.\s*PROFESIONAL QUE ATIENDE\s*([^\n]+)")
    data["profesional_linea"] = prof_line

    return data, diagnosticos, plan_items

# --------- 3) Guardar a Excel ----------
def to_excel(data: dict, diagnosticos: list, plan_items: list, out_xlsx: str):
    # Hoja 1: Resumen (una fila)
    df_resumen = pd.DataFrame([data])

    # Hoja 2: Diagnósticos
    df_dx = pd.DataFrame(diagnosticos) if diagnosticos else pd.DataFrame(columns=["codigo","descripcion"])

    # Hoja 3: Plan
    df_plan = pd.DataFrame({"plan_item": plan_items}) if plan_items else pd.DataFrame({"plan_item":[]})

    with pd.ExcelWriter(out_xlsx, engine="openpyxl") as writer:
        df_resumen.to_excel(writer, index=False, sheet_name="Resumen")
        df_dx.to_excel(writer, index=False, sheet_name="Diagnosticos")
        df_plan.to_excel(writer, index=False, sheet_name="Plan")

# --------- 4) Main ----------
if __name__ == "__main__":
    pdf_in = "Historia_Clinica_Simulada.pdf"  # ajusta la ruta si está en otra carpeta
    if not Path(pdf_in).exists():
        raise SystemExit(f"No se encontró {pdf_in}. Coloca el PDF junto al script o ajusta la ruta.")

    text = read_pdf_text(pdf_in)
    data, dx, plan = parse_historia(text)
    out = "Historia_Clinica_Parsed.xlsx"
    to_excel(data, dx, plan, out)
    print(f"Listo: {out}")
```
