# <FONT SIZE=5 COLOR="purple"> 2. Tokenización </FONT>

- Es el proceso de dividir las cadenas de texto de un documento en piezas o unidades más pequeñas que se denominan *tokens*.

- Los *tokens* pueden ser palabras, subpalabras o incluso caracteres.

- La librería ***NLTK*** tiene funciones para hacer la tokenización de textos.

```python
# para tokenizar
texto.split()
```

```python
# para revisar la longitud de la cadena
len(texto.split())
```

```python
# librerías para procesar
import numpy as np
import pandas as pd

# expresiones regulares
import re

# librerías para graficar
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# librería de PLN en español
import nltk
from nltk.corpus import stopwords
#stopwords_sp = stopwords.words('spanish')

# Count vectorizer usando NLTK
from sklearn.feature_extraction.text     import CountVectorizer
from sklearn.feature_extraction.text     import TfidfVectorizer

# Para el algoritmo k-means
from sklearn.cluster             import KMeans

# Algunos elementos de preprocesamiento: escalamiento y selección de k en k-means
from sklearn.preprocessing      import StandardScaler, MinMaxScaler
from sklearn.metrics            import silhouette_samples, silhouette_score

# Para los algotitmos de machine learning
from sklearn.decomposition      import PCA
from sklearn.neighbors          import KNeighborsClassifier
from sklearn.neighbors          import NearestNeighbors

from sklearn.metrics            import pairwise_distances_argmin_min

#Para omitir los warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# recordemos la función de preprocedado
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub('á', 'a', texto)
    texto = re.sub('é', 'e', texto)
    texto = re.sub('í', 'i', texto)
    texto = re.sub('ó', 'o', texto)
    texto = re.sub('ú', 'u', texto)
    texto = re.sub("[^a-züéáíóúñ ]", " ", texto)
  #  texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = [palabra for palabra in texto.split() if (len(palabra)>3 and palabra not in stopwords_sp) | (palabra == "no")]
   # texto = lemm.lemmatize(texto)
    texto = " ".join(texto)
    return texto
```

```python
stopwords_sp = stopwords.words('spanish')
```

```python
comentario_limpio = [pre_procesado(comentario)  for comentario in datos.comentario]
```

```python
datos["comentario_limpio"] = comentario_limpio
```

```python
datos.head()
```

```python
# aplicar el tfidf vectorizar. Recordar: sparse matrix
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(datos.comentario_limpio)
tfidf_matrix = pd.DataFrame(tfidf.toarray(), columns = tfidf_vect.get_feature_names_out())
tfidf_matrix
```

```python
# veamos el método del codo
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()                                          # definimos el modelo
visualizer = KElbowVisualizer(model, k=(1,10))            # definimos el visualizador
visualizer.fit(tfidf_matrix)                              # Entrenamos con los datos
visualizer.show()
```

```python
k = 4
model = KMeans(n_clusters=k,
               random_state=42)
model.fit(tfidf_matrix)
```

```python
datos['cluster'].value_counts(normalize=True)
```

```python
# pegamos los clusters
datos["cluster"] = model.labels_
datos[["comentario" , "cluster"]]
```

```python
vecinos = NearestNeighbors(n_neighbors=5, metric='euclidean').fit(tfidf_matrix.values)

freq_cluster = datos["cluster"].value_counts()
freqrel_cluster = datos['cluster'].value_counts(normalize=True)

centroides = model.cluster_centers_

palabras = tfidf_vect.get_feature_names_out()
order_centroides = centroides.argsort()[:, ::-1]

for i in range(k):
    print(f"Cluster # {i}")
    print(f"Tiene {freq_cluster[i]} registros ({freqrel_cluster[i]:.2%} de los datos)")
    print()

    print("Palabras más representativas")
    for ind in order_centroides[i][:20]:
        print(palabras[ind], end=" ")
    print()
    print()

    print(f"Documentos más representativas")
    for vecino in datos.iloc[vecinos.kneighbors([centroides[i]])[1][0]].comentario.values:
        print(f" *** {vecino}")
        print()

    print()
    print("********************************************")
    print()
```

# <FONT SIZE=5 COLOR="purple"> 2. LIBRERÍAS PARA PROCESAMIENTO DE LENGUAJE NATURAL PLN </FONT>

Actualmente, existen diferentes librería que se usan para hacer procesamiento de texto y aplicar metodologías de procesamiento de lenguaje natural en general (PLN)

- **NLTK** (https://www.nltk.org/): Es una librería desarrollada por Steven Bird y Edward Loper para el Procesamiento de Lenguaje Natural, principalmente en inglés, que tiene herramientas para trabajar con: corpus, recursos léxicos, algoritmos de procesamiento de PLN, etc.

- **SpaCy** (https://spacy.io/): Es una librería para el NLP incorpora funcionalidades como Tokenización, Lematización, PoS, NER, etc. en varios idiomas. A diferencia de NLTK que tienen fines de caracter didáctico, SpaCy es más aplicado en la solución de problemas reales.

- **Gensim** (https://radimrehurek.com/gensim/) Es una librería muy útil para vectorización de textos, topic modeling, LDA, etc. Preprocesamiento, entre otras aplicaciones y procesos.
