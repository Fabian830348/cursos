1. Media o promedio: **.mean()**

- Es el "equilibrio" de los datos.
- Se usa para variables cuantitativas.

- Es sensible a datos atípicos. Por está razón se debe tener cuidado con su interpretación.

````python
.mean()
````

2. Mediana: **.median()**

- Es el valor que divide los datos en dos partes iguales (50%-50%)
- Aplica para variables cuantitativas
- Puede o no estar en el conjunto de datos. Depende si se tiene un número par o impar de elementos.
- No se ve afectada por la existencia de datos atípicos. Por está razón algunas veces es usada para imputar valores atípicos.

````python
.median()
````

3. Moda: **.mode()**

- Es el valor que más se repite. Es decir el que tiene mayor frecuencia
- Aplica para variables cualitativas y cuantitativas
- UN conjunto puede no tener o tener múltiples modas.

## <FONT SIZE=4 COLOR="green"> 7.3. BOXPLOT (GRÁFICO DE CAJAS Y BIGOTES) </FONT>

- El gráfico de caja, boxplot del inglés, fue ideado por Jhon Tukey en 1977 y es una representación estadística que resalta principalmente aspectos de la distribución de los datos u observaciones por medio de las medidas de posición.

- Los diagramas de caja brindan información valiosa que no se observa en un histograma como son los cuartiles.

A continuación a describir cada uno de los elementos que tiene un boxplot

**La caja rectangular** : el gráfico contiene una caja rectangular. Es importante indicar que esta caja no tiene coordenadas cartesianas en el eje $X$  y las alturas en el eje $y$ no representa una escala de frecuencias. Las líneas horizontales en el diagrama de cajas describen el primer y tercer cuartil. $Q_1$ y $Q_3$ respectivamente.

**Mediana** : es una línea que se muestra dentro de la caja rectangular y hace referencia al valor que divide los datos en dos partes iguales. Recuerde que la mediana es igual a $Q_2$.

**Bigotes** : son líneas que salen a los costados de la caja y se usan como referencia para ubicar las observaciones que estan por fuera del $50 \%$ central de distribucion. Los bigotes nos dan información sobre la existencia de valores atípicos (*outliers*).

````python
# boxplot clásico
px.box(hospital.EDAD)
````

Podemos comenzar a hacer gráficos más eleborados que relacionen más variables

````python
px.box(hospital,                                                # nombre de los datos
       x = "HOSPITALIZACION",                                    # variable cualitativa o categórica
       y = "EDAD",                                              # variable cuantitativa
       title = "Boxplot variable EDAD vs HOSPITALIZACIÓN",      # título del gráfico
       color = "HOSPITALIZACION"                                # color por la variable categórica
       )
````

````python
px.box(hospital,                                                                # nombre de los datos
       y = "EDAD",                                                              # variable cuantitativa
       x = "ENF. CRONICA PULMONAR OBSTRUCTIVA",                                 # variable cualitativa o categórica
       title = "Boxplot variable EDAD vs ENF. CRONICA PULMONAR OBSTRUCTIVA",    # título del gráfico
       color = "ENF. CRONICA PULMONAR OBSTRUCTIVA"                              # color por la variable categórica
        )
````

# <FONT SIZE=5 COLOR="green"> 8. Adicional: Acceder a los elementos del DataFrame </FONT>

Para acceder a los elementos, o a las filas o columnas del DataFrame podemos usar:

- ***iloc*** : para acceder por la posición
- ***loc.*** : para acceder por el nombre

Si usamos la primera instrucción

- ***df.iloc[i,j]*** : elemento de la fila i y columna j.

- ***df.iloc[i]*** : devuelve la fila i

- ***df.iloc[lista i , lista j]*** : devuelve un data.frame con las filas en lista i y columnas en columna j.

- ***df.iloc[:, j]*** : devuelve toda columna j.

Si usamos la instrucción **.loc**

- ***df.loc[nombre fila , nombre columna]***: es lo mismo anterior pero con los nombres.

- ***df[nombre columna]*** : devuelve las columnas en name columnas

- ***df.columna*** : otra forma para acceder a una columna que en el nombre no tenga espacios en blanco.

````python
# las columnas 2-5
aptos.iloc[: , 2:6]
````

````python
# para seleccionar la segunda y cuarta columna
aptos.iloc[: , [1,3]]
````
