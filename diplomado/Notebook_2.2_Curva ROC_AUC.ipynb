# <FONT SIZE=5 COLOR="purple"> 1. Curva ROC (Receiver operating characteristic) </FONT>

- *ROC* es el acr√≥nimo de Receiver Operating Characteristics (caracter√≠stica operativa del receptor) y fue desarrollada en el campo de la ingenier√≠a electr√≥nica para medir la detecci√≥n de objetos enemigos en el campo de batalla usando los radades durante la segunda guerra mundial. Fue la base para la TDS-Teor√≠a de Detecci√≥n de Se√±ales.

- Sus aplicaciones trascendieron m√°s all√° del campo militar en √°reas como la medicina, la epidemiolog√≠a y en la investigaci√≥n, etc.

- Ahora, la curva ROC se aplica para evaluar algoritmos de *machine learning* y hacer comparaci√≥n del rendimiento de estos.

- Formalmente, la ***curva ROC*** es la representaci√≥n gr√°fica de la raz√≥n o proporci√≥n de verdaderos positivos (TPR = True positive rate) frente a la raz√≥n o proporci√≥n de falsos positivos (FPR = False positive rate) seg√∫n se var√≠a el umbral de discriminaci√≥n (valor a partir del cual decidimos que un caso es un positivo).

- La curva *ROC* nos dice qu√© tan bueno puede distinguir el modelo entre dos cosas, por ejemplo, si un paciente tiene c√°ncer o no. Mejores modelos pueden distinguir con precisi√≥n entre los dos, mientras que un modelo pobre tendr√° dificultades para distinguir entre los dos.

Supongamos que tenemos un modelo que predice si un paciente tiene c√°ncer o no. Teniendo en cuenta la matriz de confusi√≥n:

<center><img src="https://github.com/Fabian830348/cursos/blob/master/ROC/roc2.png?raw=true" alt="centered image" width="500" height="400"></center> <center><figcaption> <FONT SIZE=1 COLOR="black"> Fuente: Elaboraci√≥n propia   </FONT> <figcaption></center>

- Los valores positivos por encima del umbral ser√°n *verdaderos positivos*
- Los valores negativos por encima del umbral ser√°n *falsos positivos*; ya que se predicen incorrectamente como positivos.
- Los valores negativos por debajo del umbral ser√°n *verdaderos negativos*.
- Los valores positivos por debajo del umbral ser√°n *falsos negativos*, ya que se pronostican incorrectamente como negativos.

Recordemos dos f√≥rmulas de la matriz de confusi√≥n:  *sensibilidad* y *especificidad*.

La ***sensibilidad o recall***, es la proporci√≥n de pacientes que se identificaron correctamente por tener c√°ncer, es decir verdadero positivo, sobre el n√∫mero total de pacientes que realmente tienen la enfermedad.

$$recall = \dfrac{VP}{VP+FN}$$

Por su parte, ***especificidad*** es la proporci√≥n de pacientes que se identificaron correctamente por no tener c√°ncer, verdadero negativo, sobre el n√∫mero total de pacientes que no tienen la enfermedad.

$$specificity = \dfrac{VN}{VN+FP}$$

Si movemos el umbral, los valores de $VN, VP, FN, FP$ cambiar√°n y tendr√≠amos clasificaciones diferentes:

<br>

**Relaci√≥n entre *recall* y *specificity* **

- Entre mayor sea el *recall* (sensibilidad) menor ser√° la especificidad, en otras palabras, guardan una relaci√≥n inversa.

- La curva ROC es la gr√°fica de $recall$ contra

$$1- especificidad = \dfrac{FP}{VN+FP}$$

# <FONT SIZE=5 COLOR="Green"> 1.2 AUC : √Årea bajo la curva ROC </FONT>

- En el campo de machine learning el **√°rea bajo la curva ROC** (***AUC***) es un instrumento para evaluar modelos de clasificaci√≥n.

- El AUC esta relacionado con el √≠ndice de Gini:

$$Gini = 2 \times AUC -1$$

- Este puntaje nos da una buena idea de qu√© tan bien funciona el modelo.

- Un modelo ser√° mejor en cuanto su *AUC* este m√°s cercano a $1$. Es decir, que la curva ROC sea m√°s parecida a $y=1$.

## <FONT SIZE=4 COLOR="bronw"> 2.1 Importar las librer√≠as </FONT>

```python
# librer√≠as para c√°lculos e importar datos
import numpy as np
import pandas as pd

# librer√≠as para graficar
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# entrenamiento y prueba y para escalar
from sklearn.model_selection    import train_test_split
from sklearn.preprocessing      import StandardScaler
from sklearn.preprocessing      import MinMaxScaler

# Binarizar las variables categ√≥ricas
from sklearn.preprocessing import label_binarize

# generaci√≥n del modelo de regresi√≥n log√≠stica
from sklearn.linear_model       import LogisticRegression

# m√©tricas de evaluaci√≥n y las matrices de confusi√≥n
from sklearn                    import metrics
from sklearn.metrics            import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics            import confusion_matrix, classification_report, ConfusionMatrixDisplay


# selecci√≥n de caracter√≠sticas
from sklearn.feature_selection  import RFE

# b√∫squeda en grilla.
from sklearn.model_selection    import GridSearchCV

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve


# Para omitir los warnings
import warnings
warnings.filterwarnings("ignore")
```

## <FONT SIZE=4 COLOR="bronw"> 2.2 Importar los datos </FONT>

Vamos a ejecutar el modelo de la clase anterior.

```python
url ="https://raw.githubusercontent.com/Fabian830348/Bases_Datos/master/cancer_clasificacion.csv"
cancer=pd.read_csv(url, sep=";")
cancer
```

## <FONT SIZE=4 COLOR="bronw"> 2.3 Generamos y evaluamos el modelo </FONT>

En un solo bloque vamos a aplicar el modelo de regresi√≥n log√≠stica del notebook 3. Los pasos son:

- Definir las variables predictoras y de respuesta

- Codificar las categorias de la variable de respuesta

- Partir en conjuntos de entrenamiento y prueba.

- Escalar los datos. En este caso usaremos ***MinMax***

- Generar el modelo de regresi√≥n log√≠stica

- Entrenar el modelo

- Generar la matriz de confusi√≥n para su evaluaci√≥n.

```python
# definimimos los predictores y la variable de respuesta
X = cancer.iloc[:, 2:-1]
y = cancer["diagnosis"]

# binarizamos la variable. Codificamos a 0 y 1.
y = label_binarize(y, classes=["B", "M"])

# seleccionamos el conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=0)

# escalamos los datos. Tambi√©n se puede usar StandardScaler
escaladorX = MinMaxScaler()
X_train = escaladorX.fit_transform(X_train)
X_test  = escaladorX.transform(X_test)

# entrenamos el clasificador
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X_train, y_train)

# hacemos las predicciones en el conjunto de train y test.
y_pred_train = log_reg.predict(X_train)
y_pred_test  = log_reg.predict(X_test)

# calculamos las matrices de confusi√≥n para train y test
cm1 = confusion_matrix(y_train, y_pred_train)
cm2 = confusion_matrix(y_test, y_pred_test)

# Crear una figura con dos subgr√°ficos
fig, axes = plt.subplots(1, 2, figsize=(7, 3))

# Graficar la primera matriz de confusi√≥n
sns.heatmap(cm1,
            annot=True,
            annot_kws = {'size':10},
            cmap='Blues',
            fmt="g",
            ax=axes[0])
axes[0].set_title('Matriz de Confusi√≥n Train Log_reg')
axes[0].set_xlabel('Predicci√≥n')
axes[0].set_ylabel('Real')

# Graficar la segunda matriz de confusi√≥n
sns.heatmap(cm2,
            annot=True,
            annot_kws = {'size':10},
            cmap='Blues',
            fmt="g",
            ax=axes[1])
axes[1].set_title('Matriz de Confusi√≥n Test Log_reg')
axes[1].set_xlabel('Predicci√≥n')
axes[1].set_ylabel('Real')

# Mostrar el gr√°fico
plt.tight_layout()
plt.show()

print("Reporte para train\n \n", classification_report(y_train,y_pred_train))
print("Reporte para test\n \n", classification_report(y_test,y_pred_test))
```

## <FONT SIZE=4 COLOR="bronw"> 2.4 Generamos la curva ROC y AUC </FONT>

Ahora vamos a calcular la curva *ROC* y el AUC

- Para generar el √°rea bajo la curva, usamos:

$$roc\_auc\_score(y\_train, log\_reg.predict(X\_train)$$

```python
# calculamos el √°rea bajo la curva ROC: AUC
log_roc_auc = roc_auc_score(y_train, log_reg.predict(X_train))
log_roc_auc
```

Ahora, calculamos las probabilidades de predicci√≥n de cada elementos en el conjunto de entrenamiento.

- La primera componente es la $P(y=0 \mid X=x_0)$ y la segunda componente $P(y=1 \mid X=x_0)$, donde $x_0$ es el valor de las variables predictoras en cada registro del conjunto de datos de entrenamiento.

Veamos los 5 primeros registros de las instrucciones

- *modelo.predict* : predicciones
- *modelo.predict_proba* : probabilidades de predicci√≥n.

```python
print(log_reg.predict(X_train)[0:5])
print(log_reg.predict_proba(X_train)[0:5])
```

```python
# veamos la informaci√≥n de la curva roc
# esta nos da tres valores: FPR , TPR y el umbral
roc_curve(y_train, log_reg.predict_proba(X_train)[:,1])
```

```python
# calculamos la tasa de falsos positivos y verdaderos positivos
fpr, tpr, thresholds = roc_curve(y_train,
                                log_reg.predict_proba(X_train)[:,1])
# los colocamos como un DataFrame
df=pd.DataFrame(np.transpose(np.array([fpr, tpr,thresholds])), columns = ["FPR", "TPR", "Umbrales"])
df.head(10)
```

Ahora, hacemos la curva *ROC*

```python
# calculamos la tasa de falsos positivos y verdaderos positivos
fpr, tpr, thresholds = roc_curve(y_train, log_reg.predict_proba(X_train)[:,1])

# detalles de la gr√°fica
plt.figure()
plt.plot(fpr, tpr, label='Regresi√≥n Log√≠stica (area = %0.2f)' % log_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (fpr)')
plt.ylabel('True Positive Rate (tpr)')
plt.title("Curva ROC")
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

Observamos que la curva ROC r√°pidamente crece hacia $y=1$ y esto es coherente con el √°rea bajo la curva AUC que tiene un valor del 0.90.

# <FONT SIZE=5 COLOR="Purple"> 3. Selecci√≥n autom√°tica de Caracter√≠sticas **RFE**</FONT>

- La eliminaci√≥n recursiva de caracter√≠sticas (***RFE***) es un m√©todo de selecci√≥n de caracter√≠sticas que se ajusta a un modelo y elimina la caracter√≠stica (o caracter√≠sticas) m√°s d√©biles hasta que se alcanza el n√∫mero especificado de caracter√≠sticas.

- Las caracter√≠sticas se clasifican seg√∫n los atributos del modelo ***coef_*** y ***feature_importances_***, al eliminar de forma recursiva una peque√±a cantidad de caracter√≠sticas por bucle.

- *RFE* intenta eliminar las dependencias y la colinealidad que puedan existir en el modelo.

- *RFE* requiere que se mantenga una cantidad espec√≠fica de caracter√≠sticas.

- Para encontrar el n√∫mero √≥ptimo de caracter√≠sticas, se utiliza la validaci√≥n cruzada con RFE para calificar diferentes subconjuntos de caracter√≠sticas y seleccionar la colecci√≥n de caracter√≠sticas con mejor puntuaci√≥n.

- El ***RFECV*** visualizador traza la cantidad de caracter√≠sticas en el modelo junto con su puntuaci√≥n de prueba de validaci√≥n cruzada y su variabilidad y visualiza la cantidad de caracter√≠sticas seleccionadas.

El algoritmo [RFE](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html)

 Recursive feature elimination (eliminaci√≥n recursiva de variables), el usa la deviance del modelo


$$\left(-2\sum_{i=1}^n \left\{Y_i \ln(p_i)+(1-Y_i)\ln(1-p_i)\right\}\right)+K$$

En R se usar√≠a el AIC (criterio de informaci√≥n de Akaike)

```python
from yellowbrick.model_selection import RFECV
# se define el RFE-CV en el modelo de regresi√≥n log√≠stica.
visualizer = RFECV(log_reg)
# se entrena con los conjuntos de entrenamiento
visualizer.fit(X_train, y_train)
# se gr√°fica
visualizer.show()
```

```python
# para visualizar el ranking
print(visualizer.ranking_)
# para ver las variables en ese ranking
cancer.iloc[:, 3:].columns[visualizer.ranking_ == 1]
```

```python
# Le indico que me seleccione las 5 variables m√°s importantes
rfe=RFE(log_reg, n_features_to_select=5)
rfe.fit(X_train, y_train)
rfe.ranking_
X.columns[rfe.ranking_==1]
```

Podemos volver a ejecutar el modelo con esas 5 variables

```python
# definimimos los predictores y la variable de respuesta
X5 = X[X.columns[rfe.ranking_==1]]
y = cancer["diagnosis"]

# binarizamos la variable. Codificamos a 0 y 1.
y = label_binarize(y, classes=["B", "M"])

# seleccionamos el conjunto de entrenamiento y prueba
X5_train, X5_test, y_train, y_test = train_test_split(X5,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=0)

# escalamos los datos. Tambi√©n se puede usar StandardScaler
escaladorX = MinMaxScaler()
X_train = escaladorX.fit_transform(X5_train)
X_test  = escaladorX.transform(X5_test)

# entrenamos el clasificador
log_reg = LogisticRegression(random_state=0)
log_reg.fit(X5_train, y_train)

# hacemos las predicciones en el conjunto de train y test.
y_pred_train = log_reg.predict(X5_train)
y_pred_test  = log_reg.predict(X5_test)

# calculamos las matrices de confusi√≥n para train y test
cm1 = confusion_matrix(y_train, y_pred_train)
cm2 = confusion_matrix(y_test, y_pred_test)

# Crear una figura con dos subgr√°ficos
fig, axes = plt.subplots(1, 2, figsize=(7, 3))

# Graficar la primera matriz de confusi√≥n
sns.heatmap(cm1,
            annot=True,
            annot_kws = {'size':10},
            cmap='Blues',
            fmt="g",
            ax=axes[0])
axes[0].set_title('Matriz de Confusi√≥n Train Log_reg')
axes[0].set_xlabel('Predicci√≥n')
axes[0].set_ylabel('Real')

# Graficar la segunda matriz de confusi√≥n
sns.heatmap(cm2,
            annot=True,
            annot_kws = {'size':10},
            cmap='Blues',
            fmt="g",
            ax=axes[1])
axes[1].set_title('Matriz de Confusi√≥n Test Log_reg')
axes[1].set_xlabel('Predicci√≥n')
axes[1].set_ylabel('Real')

# Mostrar el gr√°fico
plt.tight_layout()
plt.show()

print("Reporte para train\n \n", classification_report(y_train,y_pred_train))
print("Reporte para test\n \n", classification_report(y_test,y_pred_test))
```
