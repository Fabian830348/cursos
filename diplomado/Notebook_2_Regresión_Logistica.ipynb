# <FONT SIZE=5 COLOR="purple"> 1. Explicación del Modelo </FONT>

- Usamos regresión logística para estudiar la probabilidad de una variable categórica cualitativa (por ejemplo, dicotómica)  en función de una o más variables cuantitativas.

- En otras palabras, la regresión logística no permite clasificar directamente, pero si permite determinar una probabilidad de pertenecer a un grupo determinado de la variable categórica. Y luego, de acuerdo con esa probabilidad, tomar una decisión de asignación.

Se puede emplear la regresión logística por ejemplo:

- La probabilidad de estar empleado en función de carrera, nivel de inglés, edad.

- La probabilidad de que alguien compre un producto en función del ingreso, estrato, género.

- La probabilidad de pagar una deuda en funcion de otras deudas, ingreso, estado civil.

### Motivación al modelo

Supongamos que tenemos una variable categórica con dos clases, que las llamaremos $\{0, 1\}$. Entonces, si tratamos de buscar un modelo lineal de la forma

$$y = \beta_0 +\beta_1 x_1 +\epsilon$$

donde $y$ es una variable binaria, no sería un buen modelo ya que la parte izquierda de la expresión es continua. Luego, podríamos pensar en no tener una relación directa, sino vía la probabilidad, así:

$$P(y_i=1 \mid X = x_i)= \beta_0+\beta_1x_i$$

Cuando intentamos usar mínimos cuadrados, es fácil ver que el modelo no funciona ya que predecimos valores mayores que 1 y menores que 0.

<center><img src="https://github.com/Fabian830348/cursos/blob/0da91b828112d77dbddb387cbb7043ab8a78f252/Imagen/log1.png?raw=true" alt="centered image" width="500" height="400"></center>

podemos observar que este modelo no se ajusta a los datos. Más aún. Si tratamos de predecir un valor, como $x=1.1$, al tratarse de una probabilidad, este valor no tiene sentido, pues es negativo.

Entonces, buscaremos una función cumpla lo siguiente

- $P(y_i=1 \mid X = x_i) = f(\beta_0 + \beta_1 x_i)$

- $P(y_i=1 \mid  x_i)$ sea creciente en $x$ para $\beta_1>0$.

- $P(y_i=1 \mid  x_i)$ sea decreciente en $x$ para $\beta_1<0$.

- $0 \leq P(y_i=1 \mid  x_i) \leq 1$ para todo valor de $x_i$.

Se puede ver rápidamente que la función debería tener una forma de S. A esta función se le denomina función sigmoide y es la base del modelo logístico

$$f(x)= \dfrac{e^x}{1+e^x}$$

La gráfica de esta función es:

<center><img src="https://github.com/Fabian830348/cursos/blob/0da91b828112d77dbddb387cbb7043ab8a78f252/Imagen/log2.png?raw=true" alt="centered image" width="500" height="400"></center>

Así pues, tenemos una expresión de la forma

$$P(y_i=1 \mid  X = x_i)= \dfrac{e^{\beta_0+\beta_1x_i}}{1+e^{\beta_0+\beta_1x_i}}$$

Recordemos que

- $P(y_i=y_0 \mid   X = x_i)$ es la probabilidad de que la variable cualitativa $y$ adquiera el valor $y_0$ (sin perdida de generalidad tomamos uno, $y_0=1$) dado que el predictor $X$ toma el valor de $x_i$

Vamos a llamar $U=P(y_i=1 \mid  x_i)$, entonces

$$ \begin{align*} U & = \dfrac{e^{\beta_0+\beta_1x_i}}{1+e^{\beta_0+\beta_1x_i}} \\ ({1+e^{\beta_0+\beta_1x_i}} )U & = e^{\beta_0+\beta_1x_i} \\U +e^{\beta_0+\beta_1x_i}U & = e^{\beta_0+\beta_1x_i} \\ e^{\beta_0+\beta_1x_i}U-e^{\beta_0+\beta_1x_i} & = -U \\ e^{\beta_0+\beta_1x_i}(U-1)& = -U \\ e^{\beta_0+\beta_1x_i} & = \dfrac{-U}{U-1} \\ \beta_0+\beta_1x_i & = \ln \left ( \dfrac{U}{1-U}  \right ) \end{align*}    $$


$$\beta_0+\beta_1x_i= \ln \left ( \dfrac{P(y_i=1 \mid  x_i)}{1-P(y_i=1 \mid  x_i)}  \right )$$

A esta última expresión se le conoce como el $LOG$ de $Odds$.

Para interpretar este logaritmo tengamos en cuenta lo siguiente. Si la probabilidad de que suceda un evento es $0.7$, entonces la probabilidad de que no suceda el evento es $1-0.7$, es decir $0.3$. Los Odds o razón de probabilidad de que se cumpla el evento, es el cociente entre la probabilidad de que sea verdadero entre la que sea falso.

```python
# grafica en python de la función logística
import numpy as np
import matplotlib.pyplot as plt

# Crear un rango de valores para x
x = np.linspace(-10, 10, 400)

# Calcular los valores de y usando la función logística
y = np.exp(x) / (1 + np.exp(x))

# Graficar la función logística
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Función Logística', color='blue')
plt.title('Función Logística')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

Para el contexto del problema revisar el notebook 2.

```python
diabetes= pd.read_csv("https://raw.githubusercontent.com/Fabian830348/Bases_Datos/master/diabetes.csv")
```

```python
# cambiar el nombre de la variable objetivo
diabetes.rename(columns = {'Outcome':'resultado'}, inplace = True)
```

```python
diabetes.head()
```

```python
# las variables predictoras
X = diabetes.drop("resultado", axis=1)
# la variable objetivo
y = diabetes["resultado"]
```

```python
# conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,                        # variables predictoras
                                                    y,                        # variable de respuesta
                                                    random_state = 0,         # semilla para que al ejecutar siempre de igual
                                                    test_size = 0.3)          # tamaño del conjunto de prueba
```

```python
# escalar los datos
escalar = StandardScaler()
# lo aplicamos a los conjuntos
X_train_s = escalar.fit_transform(X_train)
X_test_s = escalar.transform(X_test)
```

```python
# definir el modelo y entrenarlo
Log_reg = LogisticRegression(random_state=0)
Log_reg.fit(X_train_s, y_train)
```

```python
# calcular las predicciones
y_pred_log = Log_reg.predict(X_test_s)
```

```python
# matrix de confusión
print("Matriz de Confusión \n\n", confusion_matrix(y_test, y_pred_log), "\n")
```

```python
#métricas
metrics=["accuracy", "recall" , "precision", "f1"]
values = [accuracy_score(y_test,y_pred_log),
          recall_score(y_test,y_pred_log),
          precision_score(y_test,y_pred_log),
          f1_score(y_test,y_pred_log)]
print("Métricas \n", pd.DataFrame({"metrics": metrics , "values" : values}), "\n")
```

```python
# Resumen de las métricas
print("Resumen evaluación \n", classification_report(y_test, y_pred_log))
```

```python
url = "https://raw.githubusercontent.com/Fabian830348/Bases_Datos/master/compras.csv"
```

```python
#Importar datos
datos = pd.read_csv(url)
datos.head(8)
```

```python
# Cargar los datos y eliminar la columna 'Identificador'
df = datos.drop(columns=['Identificador'])

# Convertir la variable categórica 'Sexo' en variable numérica (one-hot encoding)
#df = pd.get_dummies(df, columns=['Sexo'], drop_first=True)
#df['Sexo'] = df['Sexo'].replace({"Hombre" : 1 , "Mujer " : 0})
# Convertir la variable objetivo 'Compra' en 0 y 1
df['Sexo'] = df['Sexo'].apply(lambda x: 1 if x == 'Hombre' else 0)

# Separar variables predictoras (X) y la variable objetivo (y)
X = df.drop(columns=['Compra'])  # Variables predictoras (Edad, Salario, Sexo)
y = df['Compra']  # Variable a predecir

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar dos datos
escalar = StandardScaler()
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)

# Crear el modelo de regresión logística
model = LogisticRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones sobre el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Mostrar los resultados
print(f"Exactitud del modelo: {accuracy}")
print("Matriz de confusión:")
print(conf_matrix)
```

Recordemos

1. Los **Odds** o razones de probabilidad

$$Odds = \dfrac{P(evento)}{1-P(evento)}$$

Si *evento* = "hacer una compra" y la probabilidad de hacer una compra es $0.7$ $(70\%)$ los odds son:

$$Odds = \dfrac{0.7}{1-0.7} = \dfrac{0.7}{0.3} = 2.33$$

es decir, la probabilidad de hacer una compra es 2.33 veces mayores que las probabilidades de que no ocurra.

2. Los Odds Ratio (razón de odds) indica cómo cambia la probabilidad de que ocurra un evento cuando la variable aumenta en una unidad.

$$Odds\_Ratio = \dfrac{Odds\_antes}{Odds\_despues}$$

En regresión logística es:

$$Odds\_Ratio = e^{\beta_i}$$

Interpretación del Odds Ratio:

- $Odds\_Ratio = 1$ : No hay diferencia en las odds entre los dos grupos.

- $Odds\_Ratio > 1$: El evento es más probable en el grupo 1 en comparación con el grupo 2.

- $Odds Ratio < 1$: El evento es menos probable en el grupo 1 en comparación con el grupo 2.

3. Log(odds) nos indica cómo cambia el logaritmo de las probabilidades de que ocurra un evento (en este caso, realizar una compra) cuando una variable aumenta en una unidad, manteniendo las demás constantes.

$$Log(Odds)$$

De está manera tenemos

1. **Sexo**

- Supongamos que estamos analizando el efecto del sexo en la probabilidad de realizar una compra, donde Hombre = 1 y Mujer = 0.

- El coeficiente para la variable Sexo en la regresión logística es $\beta = 0.12$ y el

- $$Odds\_Ratio = e^{0.12} = 1.128765 >1 $$

- Esto significa que Un hombre tiene un $12.8\%$ más de probabilidades de realizar una compra que una mujer, todo lo demás constante.

2. **Edad**

- Coeficiente (Log-odds): 1.933

- Esto significa que, por cada incremento de un año en la edad, el logaritmo de las probabilidades de realizar una compra aumenta en 1.933.

- Odds Ratio: 6.915

- A medida que la edad aumenta, la probabilidad de realizar una compra crece significativamente. Por cada año más de edad, las probabilidades de compra aumentan 6.9 veces.

3. **Salario**

- Coeficiente (Log-odds): 1.054

- Esto significa que, por cada unidad de incremento en el salario, el logaritmo de las probabilidades de realizar una compra aumenta en 1.054.
Odds Ratio: 2.870

- A medida que el salario aumenta, las probabilidades de realizar una compra también crecen considerablemente. Por cada incremento en el salario (dependiendo de la escala), las probabilidades de compra aumentan 2.87 veces.
